### Process categories

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
dfa=df_in[['id','categories']]
```

```python
dfa['pcategories']=dfa['categories'].apply(lambda x: ['cat_' + y.lower().replace(' ','_') for y in x])
```

```python
dfa.shape
```

```python
dfa.head()
```

```python
mlb = MultiLabelBinarizer()
```

```python
dfb = pd.DataFrame(mlb.fit_transform(dfa['pcategories']),columns=mlb.classes_, index=dfa.index)
```

```python
dfa=dfa.join(dfb)
```

```python
dfa.head()
```

```python
dfa=dfa.drop(['categories','pcategories'], axis=1)
```

```python
dfa.head()
```

```python
dfa=dfa[dfa.columns.drop(list(dfa.filter(regex='result')))]
```

```python
dfa.shape
```

```python
dfa=dfa.drop([col for col, val in dfa.sum().iteritems() if val < 4], axis=1)
```

```python
dfa.shape
```

```python
dfa.head()
```

```python
dfa.to_parquet('data/bizbuysell.categories.parquet')
```

### Process finanicals

```python
dfc=df_in[['id','financials']]
```

```python
dfc.head()
```

```python
dfc['aa']=dfc['financials'].apply(lambda x: x.lower().replace('cash flow: $','cash flow:\n$'))
```

```python
dfc['a']=dfc['aa'].apply(lambda x: x.split('\n'))
```

```python
dfc['b']=dfc['a'].apply(lambda x: list(filter(lambda y: len(y.strip()) !=0,x)))
```

```python
dfc['c']=dfc['b'].apply(lambda x: len(list(x))%2)
```

```python
dfc['d']=dfc['b'].apply(lambda x: [y.strip().lower() for y in x])
```

```python
def todict(x):
    r=dict()
    j=0
    while j < len(x)-1:
        if x[j][-1] == ':' and x[j+1][-1] != ':' and x[j+1]!='n/a':
            k=x[j]
            k=k.lower()
            k=k.replace(' ','_')
            k=k.replace('/','_')
            k=k.replace('&','_')
            k=k.replace('.','_')
            k=k.replace(':','')
            k='fin_'+k
            v=x[j+1]
            v=v.replace('$','')
            v=v.replace(',','')
            v=v.replace('*','')
            v=v.replace('per month','')
            v=v.replace('(', '-')
            v=v.replace(')', '')
            v=v.strip()
            r[k]=v
        j+=1
    return r
dfc['e']=dfc['d'].apply(todict)    
```

```python
dfc.head()
```

```python
def proc(r):
    result=dict(id=r.id)
    result.update(r.e)
    return result
dfc['f']=dfc.apply(proc, axis=1)
```

```python
dfc.head()
```

```python
dfd=pd.DataFrame(list(dfc.f))
```

```python
dfd=dfd.fillna(0)
```
