```python
import pandas as pd
```

```python
df=pd.read_json('bizbuysell.detail.lines.json',lines=True)
```

```python
df.loc[0]
```

```python
df=df[['s_id','s_category']]
df=df.dropna()
df['category_']=df['s_category'].apply(lambda x: list(str(x).lower().split('>')))
df=df.explode('category_')
df['category']=df['category_'].apply(lambda x: x.strip())
df=df[['s_id','category']]
df=df.drop_duplicates()
df.columns=['id','category']
```
